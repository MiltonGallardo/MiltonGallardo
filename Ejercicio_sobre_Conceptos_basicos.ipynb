# 1. Llama las columnas NombreEstablecimiento, NivelEscolar y SitioWeb.

# 2. Muestra el NombreEstablecimiento y DistanciaDistrito (en km) de colegios en los que DistanciaDistrito (en km) es más de 100 km.0


````python
sql = """
select EstablishmentName, [DistrictHQDistance (in Km)]
from school
where [DistrictHQDistance (in Km)]>100
"""
# Lee la consulta sql en pandas para ver los resultados
df_sql = pd.read_sql_query(sql, engine)
df_sql.head()
````

# 3. Muestra NombreEstablecimiento, NivelEscolar, DistanciaDistrito (en km) de colegios donde el nivel escolar es Junior High School.


````python
sql = """
select EstablishmentName, SchoolLevel, [DistrictHQDistance (in Km)]
from school
where SchoolLevel='PS'
group by EstablishmentName
"""
df_sql= pd.read_sql_query(sql, engine)
df_sql.head()
````

# 4. DistanciaDistrito (en km) de colegios que contengan las palabras “Junior High” en su nombre.

````python
sql = """
select EstablishmentName, [DistrictHQDistance (in Km)]
from school
where EstablishmentName like"%Junior High%"
group by EstablishmentName
"""
df_sql= pd.read_sql_query(sql, engine)
df_sql.head()
````

# 6. Muestra los nombres de los colegios de áreas urbanas. Ordena los resultados en orden alfabético inverso.

````python
sql = """
select EstablishmentName, LocatedInRuralOrUrban
from school
where  LocatedInRuralOrUrban like"%Urban%"
order by EstablishmentName DESC

"""
df_sql= pd.read_sql_query(sql, engine)
df_sql.head()
````

# 7. Repite la consulta anterior, pero renombra la columna que se muestra como “nombre”.

````python
sql = """
select EstablishmentName as nombre, LocatedInRuralOrUrban
from school
where  LocatedInRuralOrUrban like"%Urban%"
order by EstablishmentName DESC

"""
df_sql= pd.read_sql_query(sql, engine)
df_sql.head()
````
